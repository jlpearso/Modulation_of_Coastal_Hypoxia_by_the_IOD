# Subset for AS and BoB

```python
#----------------- Arabian Sea ---------------------#

ind_AS = (ds_AWG.lon < bounds_AS[1]) & (ds_AWG.lon > bounds_AS[0]) & (ds_AWG.lat < bounds_AS[3]) & (ds_AWG.lat > bounds_AS[2]) 

#subset gridded data
temp_AS = np.array(ds_AWG.temp_50_200[ind_AS])
doxy_AS = np.array(ds_AWG.doxy_50_200[ind_AS])
# nitrate_AS = np.array(ds_AWG.nitrate_50_200[ind_AS])
sla_AS = np.array(ds_SLA.daily_sla[ind_AS])
lat_AS = np.array(ds_AWG.lat[ind_AS])
lon_AS = np.array(ds_AWG.lon[ind_AS])
    
# convert to coastal
mask_AS = mask_coast(lon_AS,lat_AS,np.array(ds_AWG.coastal_mask_bounds))
temp_AS = temp_AS[mask_AS]
doxy_AS = doxy_AS[mask_AS]
# nitrate_AS = nitrate_AS[mask_AS]
sla_AS = sla_AS[mask_AS]

sla_AS = sla_AS[~np.isnan(temp_AS)]
doxy_AS = doxy_AS[~np.isnan(temp_AS)]
# nitrate_AS = nitrate_AS[~np.isnan(temp_AS)]
temp_AS = temp_AS[~np.isnan(temp_AS)]

ind = np.isnan(sla_AS) | np.isnan(doxy_AS)
doxy_AS = doxy_AS[~ind]
sla_doxy_AS = sla_AS[~ind]

ind = np.isnan(sla_AS) | np.isnan(temp_AS)
temp_AS = temp_AS[~ind]
sla_temp_AS = sla_AS[~ind]

#---------------- Western Bay of Bengal --------------------#

ind_wBoB = (ds_AWG.lon < bounds_wBoB[1]) & (ds_AWG.lon > bounds_wBoB[0]) & (ds_AWG.lat < bounds_wBoB[3]) & (ds_AWG.lat > bounds_wBoB[2]) 

#subset gridded data
temp_wBoB = np.array(ds_AWG.temp_50_200[ind_wBoB])
doxy_wBoB = np.array(ds_AWG.doxy_50_200[ind_wBoB])
# nitrate_wBoB = np.array(ds_AWG.nitrate_50_200[ind_wBoB])
sla_wBoB = np.array(ds_SLA.daily_sla[ind_wBoB])
lat_wBoB = np.array(ds_AWG.lat[ind_wBoB])
lon_wBoB = np.array(ds_AWG.lon[ind_wBoB])
    
# convert to coastal
mask_wBoB = mask_coast(lon_wBoB,lat_wBoB,np.array(ds_AWG.coastal_mask_bounds))
temp_wBoB = temp_wBoB[mask_wBoB]
doxy_wBoB = doxy_wBoB[mask_wBoB]
# nitrate_wBoB = nitrate_wBoB[mask_wBoB]
sla_wBoB = sla_wBoB[mask_wBoB]

sla_wBoB = sla_wBoB[~np.isnan(temp_wBoB)]
doxy_wBoB = doxy_wBoB[~np.isnan(temp_wBoB)]
# nitrate_wBoB = nitrate_wBoB[~np.isnan(temp_wBoB)]
temp_wBoB = temp_wBoB[~np.isnan(temp_wBoB)]

ind = np.isnan(sla_wBoB) | np.isnan(doxy_wBoB)
doxy_wBoB = doxy_wBoB[~ind]
sla_doxy_wBoB = sla_wBoB[~ind]

ind = np.isnan(sla_wBoB) | np.isnan(temp_wBoB)
temp_wBoB = temp_wBoB[~ind]
sla_temp_wBoB = sla_wBoB[~ind]


#---------------- Bay of Bengal --------------------#

ind_BoB = (ds_AWG.lon < bounds_BoB[1]) & (ds_AWG.lon > bounds_BoB[0]) & (ds_AWG.lat < bounds_BoB[3]) & (ds_AWG.lat > bounds_BoB[2]) 

#subset gridded data
temp_BoB = np.array(ds_AWG.temp_50_200[ind_BoB])
doxy_BoB = np.array(ds_AWG.doxy_50_200[ind_BoB])
# nitrate_BoB = np.array(ds_AWG.nitrate_50_200[ind_BoB])
sla_BoB = np.array(ds_SLA.daily_sla[ind_BoB])
lat_BoB = np.array(ds_AWG.lat[ind_BoB])
lon_BoB = np.array(ds_AWG.lon[ind_BoB])
    
# convert to coastal
mask_BoB = mask_coast(lon_BoB,lat_BoB,np.array(ds_AWG.coastal_mask_bounds))
temp_BoB = temp_BoB[mask_BoB]
doxy_BoB = doxy_BoB[mask_BoB]
# nitrate_BoB = nitrate_BoB[mask_BoB]
sla_BoB = sla_BoB[mask_BoB]

ind = np.isnan(sla_BoB) | np.isnan(doxy_BoB)
doxy_BoB = doxy_BoB[~ind]
sla_doxy_BoB = sla_BoB[~ind]

ind = np.isnan(sla_BoB) | np.isnan(temp_BoB)
temp_BoB = temp_BoB[~ind]
sla_temp_BoB = sla_BoB[~ind]

# nitrate_BoB = nitrate_BoB[~np.isnan(temp_BoB)]
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 3,figsize=(16,10),dpi=200)
sz = 30
fsz = 10
alpha = 0.15

cmap = get_continuous_cmap(['#034f84','#92a8d1','#b7d7e8','#f0f0f0','#f7cac9','#f7786b','#c94c4c'])


plt.rcParams.update({'font.size': 14})

##############################################################

ylim = [14,32]
xlim = [-0.45,0.45]

ax = axes[0,0]
p = ax.scatter(sla_temp_AS,temp_AS,c = 'b',edgecolors = 'b',s= sz,alpha = alpha)
ax.axvline(x=0,color = 'k',linestyle = ':')
m, b, r, p_val,_ = stats.linregress(sla_temp_AS, temp_AS)
ax.plot(sla_AS, m*sla_AS + b, 'k', linewidth = 2)
ax.set_xlim(xlim)
ax.set_ylim(ylim)
ax.set_title('Eastern Arabian Sea')
ax.set_ylabel('50-200dbar Temperature ($^\circ C$)')
add_letter(ax, 'a')
add_text(ax, 'y = ' + str(np.round(m,2)) + 'x  + ' + str(np.round(b,2)), x = 0.05,y=0.91, fontsize = fsz)
add_text(ax, 'corr coeff: ' + str(np.round(r,2)), x = 0.6,y=0.95, fontsize = fsz)
add_text(ax, 'p-value: ' + str(np.round(p_val,4)), x = 0.6,y=0.88, fontsize = fsz)

ax = axes[0,1]
p = ax.scatter(sla_temp_wBoB,temp_wBoB,c = 'b',edgecolors = 'b',s= sz,alpha = alpha)
ax.axvline(x=0,color = 'k',linestyle = ':')
m, b, r, p_val,_ =stats.linregress(sla_temp_wBoB, temp_wBoB)
ax.plot(sla_wBoB, m*sla_wBoB + b, 'k', linewidth = 2)
ax.set_xlim(xlim)
ax.set_ylim(ylim)
ax.set_title('Western Bay of Bengal')
add_letter(ax, 'b')
add_text(ax, 'y = ' + str(np.round(m,2)) + 'x  + ' + str(np.round(b,2)), x = 0.05,y=0.91, fontsize = fsz)
add_text(ax, 'corr coeff: ' + str(np.round(r,2)), x = 0.6,y=0.95, fontsize = fsz)
add_text(ax, 'p-value: ' + str(np.round(p_val,4)), x = 0.6,y=0.88, fontsize = fsz)

ax = axes[0,2]
p = ax.scatter(sla_temp_BoB,temp_BoB,c = 'b',edgecolors = 'b',s= sz,alpha = alpha)
ax.axvline(x=0,color = 'k',linestyle = ':')
m, b, r, p_val,_ =stats.linregress(sla_temp_BoB, temp_BoB)
ax.plot(sla_BoB, m*sla_BoB + b, 'k', linewidth = 2)
ax.set_xlim(xlim)
ax.set_ylim(ylim)
ax.set_title('Eastern Bay of Bengal')
add_letter(ax, 'c')
add_text(ax, 'y = ' + str(np.round(m,2)) + 'x  + ' + str(np.round(b,2)), x = 0.05,y=0.91, fontsize = fsz)
add_text(ax, 'corr coeff: ' + str(np.round(r,2)), x = 0.6,y=0.95, fontsize = fsz)
add_text(ax, 'p-value: ' + str(np.round(p_val,4)), x = 0.6,y=0.88, fontsize = fsz)

#-------------------------------------------------------------
ylim = [-50,200]
xlim = [-0.45,0.45]

ax = axes[1,0]
p = ax.scatter(sla_doxy_AS,doxy_AS,c = 'b',edgecolors = 'b',s= sz,alpha = alpha)
ax.axvline(x=0,color = 'k',linestyle = ':')
m, b, r, p_val,_ =stats.linregress(sla_doxy_AS, doxy_AS)
ax.plot(sla_AS, m*sla_AS + b, 'k', linewidth = 2)
ax.set_xlim(xlim)
ax.set_ylim(ylim)
ax.set_ylabel('50-200dbar [$O_2$] ($\mu mol/ kg$)')
add_letter(ax, 'd')
add_text(ax, 'y = ' + str(np.round(m,2)) + 'x  + ' + str(np.round(b,2)), x = 0.05,y=0.91, fontsize = fsz)
add_text(ax, 'corr coeff: ' + str(np.round(r,2)), x = 0.6,y=0.95, fontsize = fsz)
add_text(ax, 'p-value: ' + str(np.round(p_val,4)), x = 0.6,y=0.88, fontsize = fsz)
ax.set_xlabel('Sea Level Anomaly (m)')

ax = axes[1,1]
ax.scatter(sla_doxy_wBoB,doxy_wBoB,c = 'b',edgecolors = 'b',s= sz,alpha = alpha)
ax.axvline(x=0,color = 'k',linestyle = ':')
m, b, r, p_val,_ =stats.linregress(sla_doxy_wBoB, doxy_wBoB)
ax.plot(sla_wBoB, m*sla_wBoB + b, 'k', linewidth = 2)
ax.set_xlim(xlim)
ax.set_ylim(ylim)
add_letter(ax, 'e')
add_text(ax, 'y = ' + str(np.round(m,2)) + 'x  + ' + str(np.round(b,2)), x = 0.05,y=0.91, fontsize = fsz)
add_text(ax, 'corr coeff: ' + str(np.round(r,2)), x = 0.6,y=0.95, fontsize = fsz)
add_text(ax, 'p-value: ' + str(np.round(p_val,4)), x = 0.6,y=0.88, fontsize = fsz)
ax.set_xlabel('Sea Level Anomaly (m)')

ax = axes[1,2]
ax.scatter(sla_doxy_BoB,doxy_BoB,c = 'b',edgecolors = 'b',s= sz,alpha = alpha)
ax.axvline(x=0,color = 'k',linestyle = ':')
m, b, r, p_val,_ =stats.linregress(sla_doxy_BoB, doxy_BoB)
ax.plot(sla_BoB, m*sla_BoB + b, 'k', linewidth = 2)
ax.set_xlim(xlim)
ax.set_ylim(ylim)
add_letter(ax, 'f')
add_text(ax, 'y = ' + str(np.round(m,2)) + 'x  + ' + str(np.round(b,2)), x = 0.05,y=0.91, fontsize = fsz)
add_text(ax, 'corr coeff: ' + str(np.round(r,2)), x = 0.6,y=0.95, fontsize = fsz)
add_text(ax, 'p-value: ' + str(np.round(p_val,4)), x = 0.6,y=0.88, fontsize = fsz)
ax.set_xlabel('Sea Level Anomaly (m)')
#-------------------------------------------------------------
# ylim = [0,28]
# xlim = [-0.45,0.45]

# ax = axes[2,0]
# p = ax.scatter(sla_AS,nitrate_AS,c = 'b',edgecolors = 'b',s= sz,alpha = alpha)
# ax.axvline(x=0,color = 'k',linestyle = ':')
# m, b, r, p_val,_ =stats.linregress(sla_AS, nitrate_AS)
# ax.plot(sla_AS, m*sla_AS + b, 'k', linewidth = 2)
# ax.set_xlim(xlim)
# ax.set_ylim(ylim)
# ax.set_ylabel('50-200dbar [$NO_3$] ($\mu mol/ kg$)')
# ax.set_xlabel('Sea Level Anomaly (m)')
# add_letter(ax, 'g')
# add_text(ax, 'y = ' + str(np.round(m,2)) + 'x  + ' + str(np.round(b,2)), x = 0.05,y=0.91, fontsize = fsz)
# add_text(ax, 'corr coeff: ' + str(np.round(r,2)), x = 0.6,y=0.95, fontsize = fsz)
# add_text(ax, 'p-value: ' + str(np.round(p_val,4)), x = 0.6,y=0.88, fontsize = fsz)

# ax = axes[2,1]
# ax.scatter(sla_wBoB,nitrate_wBoB,c = 'b',edgecolors = 'b',s= sz,alpha = alpha)
# ax.axvline(x=0,color = 'k',linestyle = ':')
# m, b, r, p_val,_ =stats.linregress(sla_wBoB, nitrate_wBoB)
# ax.plot(sla_wBoB, m*sla_wBoB + b, 'k', linewidth = 2)
# ax.set_xlim(xlim)
# ax.set_ylim(ylim)
# ax.set_xlabel('Sea Level Anomaly (m)')
# add_letter(ax, 'h')
# add_text(ax, 'y = ' + str(np.round(m,2)) + 'x  + ' + str(np.round(b,2)), x = 0.05,y=0.91, fontsize = fsz)
# add_text(ax, 'corr coeff: ' + str(np.round(r,2)), x = 0.6,y=0.95, fontsize = fsz)
# add_text(ax, 'p-value: ' + str(np.round(p_val,4)), x = 0.6,y=0.88, fontsize = fsz)

# ax = axes[2,2]
# ax.scatter(sla_BoB,nitrate_BoB,c = 'b',edgecolors = 'b',s= sz,alpha = alpha)
# ax.axvline(x=0,color = 'k',linestyle = ':')
# m, b, r, p_val,_ =stats.linregress(sla_BoB, nitrate_BoB)
# ax.plot(sla_BoB, m*sla_BoB + b, 'k', linewidth = 2)
# ax.set_xlim(xlim)
# ax.set_ylim(ylim)
# ax.set_xlabel('Sea Level Anomaly (m)')
# add_letter(ax, 'i')
# add_text(ax, 'y = ' + str(np.round(m,2)) + 'x  + ' + str(np.round(b,2)), x = 0.05,y=0.91, fontsize = fsz)
# add_text(ax, 'corr coeff: ' + str(np.round(r,2)), x = 0.6,y=0.95, fontsize = fsz)
# add_text(ax, 'p-value: ' + str(np.round(p_val,4)), x = 0.6,y=0.88, fontsize = fsz)
```
