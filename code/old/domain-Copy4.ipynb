# Make Plot

```python
# # Plotting Full Indian Ocean

params = {'legend.fontsize': 6,
         'axes.labelsize': 11,
         'axes.titlesize': 12,
         'xtick.labelsize':10,
         'ytick.labelsize':10}

pylab.rcParams.update(params)

# limits 
cbounds = [49,104,-5,30]
cmin = 20
cmax = 160
sz = 1
cmap = plt.cm.Spectral # plt.cm.PuOr
clrs = ['navy','limegreen','deeppink','purple']
sla_clr = 'peru'
hyp_clr = 'purple'
fig, ax = plt.subplots(nrows = 1, ncols = 1,figsize=(17.8 / 2.54,4), dpi = 400,subplot_kw={'projection': ccrs.PlateCarree()})


# add in nansum cond for the columns condition to remove data where there weren't any profiles for oxygen?
mask = mask_coast(np.array(ds_WOD.lon),np.array(ds_WOD.lat),cbounds)
lat = ds_WOD.lat[mask]
lon = ds_WOD.lon[mask]
t = ds_WOD.time[mask]
temp = ds_WOD.temp[mask]
doxy = ds_WOD.doxy[mask]
pres = ds_WOD.pres
upper_doxy = doxy[:,(pres>=50) & (pres<=200)]
upper_temp = temp[:,(pres>=50) & (pres<=200)]
doxy_flag = ~np.isnan(ds_WOD.doxy_50_200[mask])
temp_flag = ~np.isnan(ds_WOD.temp_50_200[mask])

####### Bathymetry ##############################################
add_bathy(ax,cbounds)

####### Profiles ##############################################################
# Temp
ind = (~doxy_flag) & (temp_flag)
p = ax.scatter(lon[ind],lat[ind],s=sz,marker = '.',c=clrs[0],cmap=cmap,vmin=cmin,vmax=cmax,
                       transform=ccrs.PlateCarree(), label = 'Temperature Only')
print('length Temperature Only: ', len(lon[ind]))
print('length Temperature Total: ', len(lon[temp_flag]))
print('Min Time: ', np.min(t[ind]))
print('Max Time: ', np.max(t[ind]))

# Temperature and oxygen
ind = (doxy_flag) & (temp_flag)
p = ax.scatter(lon[ind],lat[ind],s=sz,marker = '.',c=clrs[1],cmap=cmap,vmin=cmin,vmax=cmax,
                       transform=ccrs.PlateCarree(), label = 'Temp. and Oxy.')
print('length Both: ', len(lon[ind]))
print('Min Time: ', np.min(t[ind]))
print('Max Time: ', np.max(t[ind]))

# Doxy
ind = (doxy_flag) & (~temp_flag)
p = ax.scatter(lon[ind],lat[ind],s=sz,marker = '.',c=clrs[2],cmap=cmap,vmin=cmin,vmax=cmax,
                       transform=ccrs.PlateCarree(), label = 'Oxygen Only')
print('length Doxy Only: ', len(lon[ind]))
print('length Doxy Total: ', len(lon[doxy_flag]))
print('Min Time: ', np.min(t[ind]))
print('Max Time: ', np.max(t[ind]))



####### Hypoxic Events ##############################################
ax.scatter(ds_HE.lon,ds_HE.lat,s=sz,marker = '.',c=hyp_clr,
                       transform=ccrs.PlateCarree(), label = 'Hypoxic Events')
#replot with better markersize - no legend label
ax.scatter(ds_HE.lon,ds_HE.lat,s=40,marker = '.',c=hyp_clr,
                       transform=ccrs.PlateCarree(), zorder = 4)
###### SLA ###################################################################

# add coastal hovemuller white squares and to legend
p = ax.scatter(ds_CL.sta_lon,ds_CL.sta_lat,s=sz,marker = '.',c=sla_clr, edgecolors=sla_clr,
                       transform=ccrs.PlateCarree(), label = 'Coastal SLA Stations')
p = ax.scatter(ds_CL.sta_lon,ds_CL.sta_lat,s=10,marker = '.',c=sla_clr, edgecolors=sla_clr,
                       transform=ccrs.PlateCarree(), zorder = 3)
# p = ax.scatter(ds_CL.sta_lon[ds_CL.loc_list],ds_CL.sta_lat[ds_CL.loc_list],s=12,
#                marker = 's',c='w', edgecolors='k',transform=ccrs.PlateCarree(), zorder = 3)

####### Land ##############################################
g = add_land(ax, bounds = cbounds,countries = True, rivers = True)
# g.xlocator = mticker.FixedLocator(np.arange(60,110,10))
g.xlabel_style = {'size': 10}
g.ylabel_style = {'size': 10}

####### Boxes ##############################################
box_bounds = [bounds_wAS,
              bounds_AS,
             bounds_wBoB,
             bounds_BoB]  #lonmin,lonmax, latmin, latmax 
# clrs = ['k','k','k','k']
clrs = ['darkslategrey','darkslategrey','darkslategrey','darkslategrey']
add_box(ax,box_bounds,clrs, linewidth = 1.5)

# box labels
plt.text(0.92, 0.90, 'EBoB', horizontalalignment='center', fontsize = 10,color = 'w',
     verticalalignment='center', transform=ax.transAxes, zorder = 4, weight='bold')

plt.text(0.63, 0.90, 'WBoB', horizontalalignment='center', fontsize = 10,color = 'w',
     verticalalignment='center', transform=ax.transAxes, zorder = 4, weight='bold')

plt.text(0.5, 0.90, 'EAS', horizontalalignment='center', fontsize = 10,color = 'w',
     verticalalignment='center', transform=ax.transAxes, zorder = 4, weight='bold')

plt.text(0.26, 0.90, 'WAS', horizontalalignment='center', fontsize = 10,color = 'w',
     verticalalignment='center', transform=ax.transAxes, zorder = 4, weight='bold')

####### Legend ##############################################
# l = ax.legend(loc='lower center',bbox_to_anchor=(.5, -.27),
#               markerscale = 12,ncol = 3, facecolor = 'lightgray');
l = ax.legend(loc='lower left',markerscale = 10,ncol = 3, facecolor = 'w',framealpha=1);

####### Bathy contour and colorbar ##############################################
add_single_bathy_cline(ax,200)
add_box(ax,[],clrs)
# cbar = add_bathy_cbar(fig,ax, lmax = 6000,pos = [0.15, .25, 0.25, 0.02])
cbar = add_bathy_cbar(fig,ax, lmax = 6000,pos = [0.13, .3, 0.22, 0.02])
cbar.set_label(label = 'm below sea level', color = 'k', fontsize = 6)
cbar.ax.tick_params(axis='x', colors='k',labelsize=5)

# do this labeled with temp, doxy, temp + doxy measurements

plt.savefig(figpath + 'domain.pdf', format='pdf', dpi = 400)
```
