# Paths and Parameters

```python
hyp_thresh = 61

IODyear_begin = '-06-01' # month-day of IOD year
IODyear_end = '-05-31' # month-day of year AFTER IOD year

figpath = '../figures/'
```

# Get Data

```python
ds_WOD = xr.open_dataset('../data/wod_coastal_processed.nc')
# ds_WOD
```

```python
ds_WOA = xr.open_dataset('../data/woa_coastal_processed.nc')
# ds_WOA
```

```python
# DMI
ds_DMI= xr.open_dataset('../data/dmi_processed.nc')
ds_DMI

posIODyears = list(np.array(ds_DMI.pos_IOD_years))
negIODyears = list(np.array(ds_DMI.neg_IOD_years))
neuIODyears = list(np.array(ds_DMI.neu_IOD_years)) 
```

# Group Into IOD Years

```python
depth = '50_200'

############################################################################################################
# Western Arabian Sea
############################################################################################################
doxy = ds_WOD['doxy_wAS_' + depth]
lat = ds_WOD.lat_wAS
lon = ds_WOD.lon_wAS
t = ds_WOD.time_wAS

# positive IOD years -------------------------------------------# 
posIODdata_wAS,_,_,posIODtime_wAS,posIODmon_wAS,_ = IOD_year_group(doxy,
                                             lat,lon,t,
                                             IODyear_begin,IODyear_end,
                                             posIODyears, region = 'wAS')

# negative IOD years -------------------------------------------# 
negIODdata_wAS,_,_,_,negIODmon_wAS,_ = IOD_year_group(doxy,
                                             lat,lon,t,
                                             IODyear_begin,IODyear_end,
                                             negIODyears, region = 'wAS')

# neutral IOD years -------------------------------------------# 
neuIODdata_wAS,_,_,_,neuIODmonwAS,_ = IOD_year_group(doxy,
                                             lat,lon,t,
                                             IODyear_begin,IODyear_end,
                                             neuIODyears, region = 'wAS')

############################################################################################################
# Eastern Arabian Sea
############################################################################################################
doxy = ds_WOD['doxy_eAS_' + depth]
lat = ds_WOD.lat_eAS
lon = ds_WOD.lon_eAS
t = ds_WOD.time_eAS

# positive IOD years -------------------------------------------# 
posIODdata_eAS,_,_,_,posIODmon_eAS,_ = IOD_year_group(doxy,
                                             lat,lon,t,
                                             IODyear_begin,IODyear_end,
                                             posIODyears, region = 'eAS')

# negative IOD years -------------------------------------------# 
negIODdata_eAS,_,_,_,negIODmon_eAS,_ = IOD_year_group(doxy,
                                             lat,lon,t,
                                             IODyear_begin,IODyear_end,
                                             negIODyears, region = 'eAS')

# neutral IOD years -------------------------------------------# 
neuIODdata_eAS,_,_,_,neuIODmon_eAS,_ = IOD_year_group(doxy,
                                             lat,lon,t,
                                             IODyear_begin,IODyear_end,
                                             neuIODyears, region = 'eAS')

############################################################################################################
# Western Bay of Bengal
############################################################################################################
doxy = ds_WOD['doxy_wBoB_' + depth]
lat = ds_WOD.lat_wBoB
lon = ds_WOD.lon_wBoB
t = ds_WOD.time_wBoB

# positive IOD years -------------------------------------------# 
posIODdata_wBoB,_,_,_,posIODmon_wBoB,_ = IOD_year_group(doxy,
                                             lat,lon,t,
                                             IODyear_begin,IODyear_end,
                                             posIODyears, region = 'wBoB')

# negative IOD years -------------------------------------------# 
negIODdata_wBoB,_,_,_,negIODmon_wBoB,_ = IOD_year_group(doxy,
                                             lat,lon,t,
                                             IODyear_begin,IODyear_end,
                                             negIODyears, region = 'wBoB')

# neutral IOD years -------------------------------------------# 
neuIODdata_wBoB,_,_,_,neuIODmon_wBoB,_ = IOD_year_group(doxy,
                                             lat,lon,t,
                                             IODyear_begin,IODyear_end,
                                             neuIODyears, region = 'wBoB')

############################################################################################################
# Eastern Bay of Bengal
############################################################################################################
doxy = ds_WOD['doxy_eBoB_' + depth]
lat = ds_WOD.lat_eBoB
lon = ds_WOD.lon_eBoB
t = ds_WOD.time_eBoB

# positive IOD years -------------------------------------------# 
posIODdata_eBoB,_,_,_,posIODmon_eBoB,_ = IOD_year_group(doxy,
                                             lat,lon,t,
                                             IODyear_begin,IODyear_end,
                                             posIODyears, region = 'eBoB')

# negative IOD years -------------------------------------------# 
negIODdata_eBoB,_,_,_,negIODmon_eBoB,_ = IOD_year_group(doxy,
                                             lat,lon,t,
                                             IODyear_begin,IODyear_end,
                                             negIODyears, region = 'eBoB')

# neutral IOD years -------------------------------------------# 
neuIODdata_eBoB,_,_,_,neuIODmon_eBoB,_ = IOD_year_group(doxy,
                                             lat,lon,t,
                                             IODyear_begin,IODyear_end,
                                             neuIODyears, region = 'eBoB')
```

# Bar Chart

```python
# Make Histograms 
# hatch_mask = min_mean > hyp_thresh
letters = ['a','b','c','d','e','f']
x = [1,2,3,4]

fig = plt.figure(figsize=(8.7 / 2.54, 4), dpi=400)
ax2 = fig.add_subplot(2,1,1); ax2.yaxis.grid(color='gainsboro');ax2.set_axisbelow(True) 
ax3 = fig.add_subplot(2,1,2); ax3.yaxis.grid(color='gainsboro');ax3.set_axisbelow(True) 

params = {'legend.fontsize': 6,
         'axes.labelsize': 6,
         'axes.titlesize': 7,
         'xtick.labelsize':6,
         'ytick.labelsize':6,
         'hatch.color': 'lightgrey'}

pylab.rcParams.update(params)

text_fz = 4
sq_sz = 20

cmap = get_continuous_cmap(['#000000','#7A378B','#42d4f4','#D3D3D3','#DcDcDc','#F0F0F0'],
                           [0,0.4,0.6,0.8,0.9,1])
cmap = plt.cm.PuOr_r
cmin = -0.15
cmax = 0.15

####################################################################################################
# Bar Charts
####################################################################################################
data_wAS = [posIODdata_wAS,negIODdata_wAS,ds_WOD.doxy_wAS_50_200,ds_WOA.doxy_wAS_50_200]
data_eAS = [posIODdata_eAS,negIODdata_eAS,ds_WOD.doxy_eAS_50_200,ds_WOA.doxy_eAS_50_200]
data_wBoB = [posIODdata_wBoB,negIODdata_wBoB,ds_WOD.doxy_wBoB_50_200,ds_WOA.doxy_wBoB_50_200]
data_eBoB = [posIODdata_eBoB,negIODdata_eBoB,ds_WOD.doxy_eBoB_50_200,ds_WOA.doxy_eBoB_50_200]

time_data_wAS = [posIODmon_wAS,negIODmon_wAS,ds_WOD.mon_wAS,ds_WOA.mon_wAS]
time_data_eAS = [posIODmon_eAS,negIODmon_eAS,ds_WOD.mon_eAS,ds_WOA.mon_eAS]
time_data_wBoB = [posIODmon_wBoB,negIODmon_wBoB,ds_WOD.mon_wBoB,ds_WOA.mon_wBoB]
time_data_eBoB = [posIODmon_eBoB,negIODmon_eBoB,ds_WOD.mon_eBoB,ds_WOA.mon_eBoB]


for pp in range(4):

    if pp==0:
        pdata = data_wAS[0]
        ndata = data_wAS[1]
        alldata = data_wAS[2]
        sdata = data_wAS[3]

        ptdata = time_data_wAS[0]
        ntdata = time_data_wAS[1]
        alltdata = time_data_wAS[2]
        stdata = time_data_wAS[3]

    elif pp==1:
        pdata = data_eAS[0]
        ndata = data_eAS[1]
        alldata = data_eAS[2]
        sdata = data_eAS[3]

        ptdata = time_data_eAS[0]
        ntdata = time_data_eAS[1]
        alltdata = time_data_eAS[2]
        stdata = time_data_eAS[3]

    elif pp==2:
        pdata = data_wBoB[0]
        ndata = data_wBoB[1]
        alldata = data_wBoB[2]
        sdata = data_wBoB[3]

        ptdata = time_data_wBoB[0]
        ntdata = time_data_wBoB[1]
        alltdata = time_data_wBoB[2]
        stdata = time_data_wBoB[3]

    else:
        pdata = data_eBoB[0]
        ndata = data_eBoB[1]
        alldata = data_eBoB[2]
        sdata = data_eBoB[3]

        ptdata = time_data_eBoB[0]
        ntdata = time_data_eBoB[1]
        alltdata = time_data_eBoB[2]
        stdata = time_data_eBoB[3]

    # O2
    x1 = pdata[~np.isnan(pdata)]
    x2 = ndata[~np.isnan(ndata)]
    x3 = alldata[~np.isnan(alldata)]
    x4 = sdata[~np.isnan(sdata)]

    x1t = ptdata[~np.isnan(pdata)]
    x2t = ntdata[~np.isnan(ndata)]
    x3t = alltdata[~np.isnan(alldata)]
    x4t = stdata[~np.isnan(sdata)]

    # get seasonal indicies
    ind_sf = (x1t == 6) | (x1t == 7) | (x1t == 8) | (x1t == 9) | (x1t == 10) | (x1t == 11)
    ind_ws = (x1t == 12) | (x1t == 1) | (x1t == 2) | (x1t == 3) | (x1t == 4) | (x1t ==5)

    x1_sf = x1[ind_sf]
    x1_ws = x1[ind_ws]

    ind_sf = (x2t == 6) | (x2t == 7) | (x2t == 8) | (x2t == 9) | (x2t == 10) | (x2t == 11)
    ind_ws = (x2t == 12) | (x2t == 1) | (x2t == 2) | (x2t == 3) | (x2t == 4) | (x2t ==5)

    x2_sf = x2[ind_sf]
    x2_ws = x2[ind_ws]

    ind_sf = (x3t == 6) | (x3t == 7) | (x3t == 8) | (x3t == 9) | (x3t == 10) | (x3t == 11)
    ind_ws = (x3t == 12) | (x3t == 1) | (x3t == 2) | (x3t == 3) | (x3t == 4) | (x3t ==5)

    x3_sf = x3[ind_sf]
    x3_ws = x3[ind_ws]
    
    ind_sf = (x4t == 6) | (x4t == 7) | (x4t == 8) | (x4t == 9) | (x4t == 10) | (x4t == 11)
    ind_ws = (x4t == 12) | (x4t == 1) | (x4t == 2) | (x4t == 3) | (x4t == 4) | (x4t ==5)

    x4_sf = x4[ind_sf]
    x4_ws = x4[ind_ws]

#     # close to hypoxic 
#     piod_hyp_c = np.round(np.sum(x1_sf<71)/len(x1_sf)*100,2)
#     niod_hyp_c = np.round(np.sum(x2_sf<71)/len(x2_sf)*100,2)
    seas_hyp_c = np.round(np.sum(x4_sf<80)/len(x4_sf)*100,2)

#     ax1.bar(x[pp]-0.2,piod_hyp_c, color='navy', align='center', width = 0.2, alpha = 0.15)
#     ax1.bar(x[pp],niod_hyp_c, color='darkorange', align='center', width = 0.2, alpha = 0.15)
    ax2.bar(x[pp]+0.2,seas_hyp_c, color='darkgreen', align='center', 
            width = 0.2, alpha = 0.15)
    
    # percents
    piod_hyp = np.round(np.sum(x1_sf<61)/len(x1_sf)*100,2)
    niod_hyp = np.round(np.sum(x2_sf<61)/len(x2_sf)*100,2)
    seas_hyp = np.round(np.sum(x4_sf<61)/len(x4_sf)*100,2)

    if pp == 0:
        ax2.bar(x[pp]-0.2,piod_hyp, color='navy', align='center', width = 0.2,
               hatch = '////')
        ax2.bar(x[pp],niod_hyp, color='darkorange', align='center', width = 0.2,
               hatch = '////')
        ax2.bar(x[pp]+0.2,seas_hyp, color='darkgreen', align='center', width = 0.2)
    else:
        ax2.bar(x[pp]-0.2,piod_hyp, color='navy', align='center', width = 0.2)
        ax2.bar(x[pp],niod_hyp, color='darkorange', align='center', width = 0.2)
        ax2.bar(x[pp]+0.2,seas_hyp, color='darkgreen', align='center', width = 0.2)
    
    ax2.text(x[pp]-0.2,1.002*piod_hyp,'%d' % len(x1_sf), ha='center', 
             va='bottom', fontsize = text_fz, color = 'gray')
    ax2.text(x[pp],1.002*niod_hyp,'%d' % len(x2_sf), ha='center', 
             va='bottom', fontsize =text_fz, color = 'gray')
    ax2.text(x[pp]+0.2,1.002*seas_hyp_c,'%d' % len(x4_sf), ha='center', 
             va='bottom', fontsize = text_fz, color = 'gray')

    ax2.set_title('Percentage of Hypoxic Profiles', 
                  pad = 8)
    ax2.set_ylabel('Summer/Fall')
    ax2.set_xticks([1,2,3,4])
    ax2.set_xticklabels([])
    ax2.set_yticks([0,10,20,30,40,50,60,70])
    ax2.set_yticklabels(['0%','10%','20%','30%','40%','50%','60%','70%'])
    ax2.set_ylim([0,75])
    
#     # close to hypoxic 
#     piod_hyp_c = np.round(np.sum(x1_ws<71)/len(x1_ws)*100,2)
#     niod_hyp_c = np.round(np.sum(x2_ws<71)/len(x2_ws)*100,2)
    seas_hyp_c = np.round(np.sum(x4_ws<80)/len(x4_ws)*100,2)

#     ax2.bar(x[pp]-0.2,piod_hyp_c, color='navy', align='center', width = 0.2, alpha = 0.15)
#     ax2.bar(x[pp],niod_hyp_c, color='darkorange', align='center', width = 0.2, alpha = 0.15)
    ax3.bar(x[pp]+0.2,seas_hyp_c, color='darkgreen', align='center', 
            width = 0.2, alpha = 0.15)

    # percents 
    
    piod_hyp = np.round(np.sum(x1_ws<61)/len(x1_ws)*100,2)
    niod_hyp = np.round(np.sum(x2_ws<61)/len(x2_ws)*100,2)
    seas_hyp = np.round(np.sum(x4_ws<61)/len(x4_ws)*100,2)
    
    if pp == 0:
        ax3.bar(x[pp]-0.2,piod_hyp, color='navy', align='center', width = 0.2, 
                label = 'Positive IOD Phase')
        ax3.bar(x[pp]-0.2,piod_hyp, color='navy', align='center', 
                width = 0.2, hatch = '////')
        ax3.bar(x[pp],niod_hyp, color='darkorange', align='center', width = 0.2, 
                label = 'Negative IOD Phase')
        ax3.bar(x[pp],niod_hyp, color='darkorange', align='center', width = 0.2, 
                hatch = '////')
        ax3.bar(x[pp]+0.2,seas_hyp, color='darkgreen', align='center', width = 0.2, 
                label = 'Annual Climatology')
    else:
        ax3.bar(x[pp]-0.2,piod_hyp, color='navy', align='center', width = 0.2)
        ax3.bar(x[pp],niod_hyp, color='darkorange', align='center', width = 0.2)
        ax3.bar(x[pp]+0.2,seas_hyp, color='darkgreen', align='center', width = 0.2)


    ax3.text(x[pp]-0.2,1.002*piod_hyp,'%d' % len(x1_ws), ha='center', 
             va='bottom', fontsize = text_fz, color = 'gray')
    ax3.text(x[pp],1.002*niod_hyp,'%d' % len(x2_ws), ha='center', 
             va='bottom', fontsize =text_fz, color = 'gray')
    ax3.text(x[pp]+0.2,1.002*seas_hyp_c,'%d' % len(x4_ws), ha='center', 
             va='bottom', fontsize = text_fz, color = 'gray')

    ax3.set_ylabel('Winter/Spring')
    ax3.set_xticks([1,2,3,4])
    ax3.set_xticklabels(['WAS','EAS','WBoB','EBoB'])
    ax3.set_yticks([0,10,20,30,40,50,60,70])
    ax3.set_yticklabels(['0%','10%','20%','30%','40%','50%','60%','70%'])
    ax3.set_ylim([0,75])

plt.legend(framealpha=1)

plt.savefig(figpath + 'hypoxia.pdf', format='pdf', dpi = 400)
```
