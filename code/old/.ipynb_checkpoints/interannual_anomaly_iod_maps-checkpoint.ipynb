```python
# import your standard packages
%run ../../global-tools/start.py

# import your local functions
sys.path.insert(1, '../local-tools/')
from local_functions import *

# make sure the figures plot inline rather than at the end
%matplotlib inline
```

```python
# CP
path_CD = '../data/CD/'
infn_CD = 'awg_cd.nc'

ds_CD= xr.open_dataset(path_CD + infn_CD)
# ds_CD
time_slice = slice('1958-06-01', '2020-05-31')
ds_CD = ds_CD.sel(time = time_slice)
ds_CD
```

```python
# get data from tigress
path = '../data/SLA/'
# infn = 'all_adt_sla_global_twosat_phy_l4_vDT2018_daily_1994_2018_processed.nc'
infn = 'sla_25km_processed.nc'

ds_SLA = xr.open_dataset(path + infn)
# ds_SLA
```

```python
# load DMI data
ds_DMI= xr.open_dataset('../data/DMI/dmi_processed.nc')
ds_DMI

posIODyears = list(np.array(ds_DMI.pos_IOD_years))
negIODyears = list(np.array(ds_DMI.neg_IOD_years))
neuIODyears = list(np.array(ds_DMI.neu_IOD_years)) 

# groupd into IOD years

IODyear_begin = '-06-01' # month-day of IOD year
IODyear_end = '-05-31' # month-day of year AFTER IOD year
```

```python
# Group into IOD years

var = ds_CD.doxy_anom_50_200_ave

binwidth = 2
bounds = [35,120,-20,30]

####################################
# average over the positive IOD years -------------------------------------------# 
posIODdata,posIODlat,posIODlon,posIODtime,posIODmon,posIODseas = IOD_year_group_AWG(var,
                                             ds_CD.lat,ds_CD.lon,ds_CD.time,
                                             IODyear_begin,IODyear_end,posIODyears)

# total
posIOD_doxy_anom_bin, posIOD_doxy_anom_bncts,latbins,lonbins = latlonbin(posIODdata,posIODlat,
                                                                               posIODlon,bounds,
                                                                               binwidth)

# s/f
ind = (posIODmon == 6)| (posIODmon == 7) | (posIODmon == 8)| (posIODmon == 9)| (posIODmon == 10)| (posIODmon == 11)
posIOD_doxy_anom_bin_sf, posIOD_doxy_anom_bncts_sf,latbins,lonbins = latlonbin(posIODdata[ind],posIODlat[ind],
                                                                               posIODlon[ind],bounds,
                                                                               binwidth)


# w/s
ind = (posIODmon == 12)| (posIODmon == 1) | (posIODmon == 2)| (posIODmon == 3)| (posIODmon == 4)| (posIODmon == 5)
posIOD_doxy_anom_bin_ws, posIOD_doxy_anom_bncts_ws,_,_ = latlonbin(posIODdata[ind],posIODlat[ind],posIODlon[ind]
                                                             ,bounds,binwidth)



# average over the negative IOD years -------------------------------------------# 
negIODdata,negIODlat,negIODlon,negIODtime,negIODmon,negIODseas = IOD_year_group_AWG(var,
                                             ds_CD.lat,ds_CD.lon,ds_CD.time,
                                             IODyear_begin,IODyear_end,negIODyears)

# total
negIOD_doxy_anom_bin, negIOD_doxy_anom_bncts,_,_= latlonbin(negIODdata,negIODlat,
                                                                         negIODlon,bounds,binwidth)

# s/f
ind = (negIODmon == 6)| (negIODmon == 7) | (negIODmon == 8)| (negIODmon == 9)| (negIODmon == 10)| (negIODmon == 11)
negIOD_doxy_anom_bin_sf, negIOD_doxy_anom_bncts_sf,_,_= latlonbin(negIODdata[ind],negIODlat[ind],
                                                                         negIODlon[ind],bounds,binwidth)

# w/s
ind = (negIODmon == 12)| (negIODmon == 1) | (negIODmon == 2)| (negIODmon == 3)| (negIODmon == 4)| (negIODmon == 5)
negIOD_doxy_anom_bin_ws, negIOD_doxy_anom_bncts_ws,_,_ = latlonbin(negIODdata[ind],negIODlat[ind],
                                                                         negIODlon[ind],bounds,binwidth)


# average over the neutral IOD years -------------------------------------------# 
neuIODdata,neuIODlat,neuIODlon,neuIODtime,neuIODmon,neuIODseas = IOD_year_group_AWG(var,
                                             ds_CD.lat,ds_CD.lon,ds_CD.time,
                                             IODyear_begin,IODyear_end,neuIODyears)

# total
neuIOD_doxy_anom_bin, neuIOD_doxy_anom_bncts,_,_ = latlonbin(neuIODdata,neuIODlat,
                                                                         neuIODlon,bounds,binwidth)

# s/f
ind = (neuIODmon == 6)| (neuIODmon == 7) | (neuIODmon == 8)| (neuIODmon == 9)| (neuIODmon == 10)| (neuIODmon == 11)
neuIOD_doxy_anom_bin_sf, neuIOD_doxy_anom_bncts_sf,_,_ = latlonbin(neuIODdata[ind],neuIODlat[ind],
                                                                         neuIODlon[ind],bounds,binwidth)

# w/s
ind = (neuIODmon == 12)| (neuIODmon == 1) | (neuIODmon == 2)| (neuIODmon == 3)| (neuIODmon == 4)| (neuIODmon == 5)
neuIOD_doxy_anom_bin_ws, neuIOD_doxy_anom_bncts_ws,_,_ = latlonbin(neuIODdata[ind],neuIODlat[ind],
                                                                         neuIODlon[ind],bounds,binwidth)



# add it in for all data
all_doxy_anom_bin, all_doxy_anom_bncts,_,_ = latlonbin(var,ds_CD.lat,ds_CD.lon,bounds,binwidth)
```

```python
var = ds_SLA.mon_sla_mon_anom

    
# average over the positive IOD years -------------------------------------------# 
posIOD_sla = IOD_year_group_SLA(var,IODyear_begin,IODyear_end,posIODyears)

all_posIOD_sla = np.nanmean(posIOD_sla,0)

ind = [5,6,7,8,9,10] # Summer/Fall - -1 for index
posIOD_sla_sf = np.nanmean(posIOD_sla [ind,:,:],0)

ind = [0,1,2,3,4,11] # Winter/Spring
posIOD_sla_ws = np.nanmean(posIOD_sla [ind,:,:],0)


# average over the positive IOD years -------------------------------------------# 
negIOD_sla = IOD_year_group_SLA(var,IODyear_begin,IODyear_end,negIODyears)

all_negIOD_sla = np.nanmean(negIOD_sla,0)

ind = [5,6,7,8,9,10] # Summer/Fall - -1 for index
negIOD_sla_sf = np.nanmean(negIOD_sla[ind,:,:],0)

ind = [0,1,2,3,4,11] # Winter/Spring
negIOD_sla_ws = np.nanmean(negIOD_sla[ind,:,:],0)

# average over the positive IOD years -------------------------------------------# 
neuIOD_sla = IOD_year_group_SLA(var,IODyear_begin,IODyear_end,neuIODyears)

all_neuIOD_sla = np.nanmean(neuIOD_sla,0)

ind = [5,6,7,8,9,10] # Summer/Fall - -1 for index
neuIOD_sla_sf = np.nanmean(neuIOD_sla[ind,:,:],0)

ind = [0,1,2,3,4,11] # Winter/Spring
neuIOD_sla_ws = np.nanmean(neuIOD_sla[ind,:,:],0)
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 3,figsize=(20,8),
                         dpi = 200,subplot_kw={'projection': ccrs.PlateCarree()})

hatch_lim = 10
cbounds = [45,100,0,30]
sz = 150
cmin = -25
cmax = 25
cmap = plt.cm.coolwarm
xx,yy = np.meshgrid(lonbins,latbins)
xx = xx.flatten()
yy = yy.flatten()

mask = mask_coast(xx,yy,[35,120,-20,30])
mask.shape

plt.rcParams.update({'font.size': 14})
levels = np.linspace(-0.15,0.15,20)

##############################################################cbounds = [35,105,10,30]
tmpxx = xx[mask]
tmpyy = yy[mask]

# posIOD s/f
ax = axes[0,0]
tmp = posIOD_doxy_anom_bin_sf
tmp = tmp.flatten()
tmp = tmp[mask]

p = ax.scatter(tmpxx,tmpyy,c = tmp,marker='s',s=sz,
                          cmap=cmap,vmin=cmin,vmax=cmax,transform=ccrs.PlateCarree())
c=ax.contour(ds_SLA.lon,ds_SLA.lat,posIOD_sla_sf, colors = 'black',
          levels = levels)

add_land(ax,cbounds)
ax.text(-0.15, 0.5, 'Positive IOD Years', va='bottom', ha='center',
        rotation='vertical', rotation_mode='anchor',
        transform=ax.transAxes, fontsize = 18)
ax.set_title('Summer/Fall')

# posIOD w/s
ax = axes[0,1]
tmp = posIOD_doxy_anom_bin_ws
tmp = tmp.flatten()
tmp = tmp[mask]

p = ax.scatter(tmpxx,tmpyy,c = tmp,marker='s',s=sz,
                          cmap=cmap,vmin=cmin,vmax=cmax,transform=ccrs.PlateCarree())
c=ax.contour(ds_SLA.lon,ds_SLA.lat,posIOD_sla_ws, colors = 'black',
          levels = levels)

add_land(ax,cbounds)
ax.set_title('Winter/Spring')

# posIOD total hatched 
ax = axes[0,2]
tmp = posIOD_doxy_anom_bin
tmp = tmp.flatten()
tmp = tmp[mask]

tmp_bncts = posIOD_doxy_anom_bncts
tmp_bncts = tmp_bncts.flatten()
tmp_bncts = tmp_bncts[mask]

p = ax.scatter(tmpxx[tmp_bncts <hatch_lim],tmpyy[tmp_bncts <hatch_lim],c = tmp[tmp_bncts <hatch_lim],marker='s',s= sz,hatch='//',
                          cmap=cmap,vmin=cmin,vmax=cmax,transform=ccrs.PlateCarree())
p = ax.scatter(tmpxx[tmp_bncts >=hatch_lim],tmpyy[tmp_bncts >=hatch_lim],c = tmp[tmp_bncts >=hatch_lim],marker='s',s= sz,
                          cmap=cmap,vmin=cmin,vmax=cmax,transform=ccrs.PlateCarree())

add_land(ax,cbounds)
ax.set_title('Full Year')


########## NEG IOD ###################
# negIOD s/f
ax = axes[1,0]
tmp = negIOD_doxy_anom_bin_sf
tmp = tmp.flatten()
tmp = tmp[mask]

p = ax.scatter(tmpxx,tmpyy,c = tmp,marker='s',s=sz,
                          cmap=cmap,vmin=cmin,vmax=cmax,transform=ccrs.PlateCarree())
c=ax.contour(ds_SLA.lon,ds_SLA.lat,negIOD_sla_sf, colors = 'black',
          levels = levels)

add_land(ax,cbounds)
ax.text(-0.15, 0.5, 'Negative IOD Years', va='bottom', ha='center',
        rotation='vertical', rotation_mode='anchor',
        transform=ax.transAxes, fontsize = 18)

# negIOD w/s
ax = axes[1,1]
tmp = negIOD_doxy_anom_bin_ws
tmp = tmp.flatten()
tmp = tmp[mask]

p = ax.scatter(tmpxx,tmpyy,c = tmp,marker='s',s=sz,
                          cmap=cmap,vmin=cmin,vmax=cmax,transform=ccrs.PlateCarree())
c=ax.contour(ds_SLA.lon,ds_SLA.lat,negIOD_sla_ws, colors = 'black',
          levels = levels)

add_land(ax,cbounds)

# negIOD total hatched 
ax = axes[1,2]
tmp = negIOD_doxy_anom_bin
tmp = tmp.flatten()
tmp = tmp[mask]

tmp_bncts = negIOD_doxy_anom_bncts
tmp_bncts = tmp_bncts.flatten()
tmp_bncts = tmp_bncts[mask]

p = ax.scatter(tmpxx[tmp_bncts <hatch_lim],tmpyy[tmp_bncts <hatch_lim],c = tmp[tmp_bncts <hatch_lim],marker='s',s= sz,hatch='//',
                          cmap=cmap,vmin=cmin,vmax=cmax,transform=ccrs.PlateCarree())
p = ax.scatter(tmpxx[tmp_bncts >=hatch_lim],tmpyy[tmp_bncts >=hatch_lim],c = tmp[tmp_bncts >=hatch_lim],marker='s',s= sz,
                          cmap=cmap,vmin=cmin,vmax=cmax,transform=ccrs.PlateCarree())

add_land(ax,cbounds)


#cbar

cbar_ax = fig.add_axes([0.925, 0.125, 0.02, 0.75])
cbar = fig.colorbar(p,cax=cbar_ax, pad=0.04, extend = 'both')
cbar.set_label('50-200dbar [$O_2$] Interannual Anomaly ($\mu mol/ kg$)')




```
