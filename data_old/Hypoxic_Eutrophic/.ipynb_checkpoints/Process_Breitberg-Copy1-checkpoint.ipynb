```python
path_AWG = '/home/jennap/projects/LRGROUP/shared_data/'
infn_AWG = 'Hypoxic_Eutrophic_Locations_Breitberg2018/hypoxic-eutrophic 2018 sites.xlsx'

df = pd.read_excel(path_AWG + infn_AWG)
df
```

```python
# subset for Indian Ocean

df = df[(df['Long ']>50) & (df['Long ']<100)& (df['Lat ']>-5)& (df['Lat ']<25)]

# add new column with numeric classifications
df['Numeric_Classification'] = [1 if x =='Hypoxic' else 0 for x in df['Classification ']] 

df
```

```python
lat = df['Lat ']
lon = df['Long ']
etype = df['Numeric_Classification']
decade = df['Decade ']
```

```python
# convert to xarray dataset
ds=xr.Dataset()

# add variables to dataset

ds["etype"]=xr.DataArray(etype,dims = ['decade'],
                     coords =[decade])

ds["lat"]=xr.DataArray(lat,dims = ['decade'],
                     coords =[decade])
ds["lon"]=xr.DataArray(lon,dims = ['decade'],
                     coords =[decade])

outfn = 'hypoxic_eutrophic_events_io.nc'
# delete if already present
if os.path.isfile(outfn):
    os.remove(outfn)

ds.to_netcdf(outfn,mode='w',format = "NETCDF4")
ds
```
